```python
fw_df['Goals_xG_Ratio'] = fw_df['Gls'] / fw_df['xG']

# Label performance
def label_performance(row):
    if row['Gls'] > row['xG']:
        return 'Overperformer'
    elif row['Gls'] < row['xG']:
        return 'Underperformer'
    else:
        return 'As Expected'

fw_df['Performance'] = fw_df.apply(label_performance, axis=1)
```

```python
fw_df = fw_df.sort_values(by='Goals_xG_Ratio', ascending=False)
overperformers = fw_df[fw_df['Performance'] == 'Overperformer']
underperformers = fw_df[fw_df['Performance'] == 'Underperformer']
```

```python
df['Min'] = pd.to_numeric(df['Min'], errors='coerce')
df['MP'] = pd.to_numeric(df['MP'], errors='coerce')
df['Gls'] = pd.to_numeric(df['Gls'], errors='coerce')
df['xG'] = pd.to_numeric(df['xG'], errors='coerce')
df['Pos'] = df['Pos'].astype(str).str.strip()
```

```python
fw_df = df[
    (df['Pos'] == 'FW') &
    (df['MP'] > 10) &
    (df['Min'] > 0) &
    (df['Gls'].notna()) &
    (df['xG'].notna()) &
    (df['xG'] > 0)
].copy()
```

```python
fw_df['Goals_xG_Ratio'] = fw_df['Gls'] / fw_df['xG']

def label_performance(row):
    if row['Gls'] > row['xG']:
        return 'Overperformer'
    elif row['Gls'] < row['xG']:
        return 'Underperformer'
    else:
        return 'As Expected'

fw_df['Performance'] = fw_df.apply(label_performance, axis=1)

# Step 4: Get top overperformer per team
top_over = fw_df[fw_df['Performance'] == 'Overperformer'] \
    .sort_values(by='Goals_xG_Ratio', ascending=False) \
    .groupby('Team') \
    .first() \
    .reset_index()[['Team', 'Player', 'Goals_xG_Ratio']] \
    .rename(columns={'Player': 'Top_Overperformer', 'Goals_xG_Ratio': 'Over_Ratio'})

# Step 5: Get top underperformer per team
top_under = fw_df[fw_df['Performance'] == 'Underperformer'] \
    .sort_values(by='Goals_xG_Ratio') \
    .groupby('Team') \
    .first() \
    .reset_index()[['Team', 'Player', 'Goals_xG_Ratio']] \
    .rename(columns={'Player': 'Top_Underperformer', 'Goals_xG_Ratio': 'Under_Ratio'})

# Step 6: Merge results
performance_leaders = pd.merge(top_over, top_under, on='Team', how='outer')
```

```python
import matplotlib.pyplot as plt

# Ensure teams are sorted for display
performance_leaders = performance_leaders.sort_values(by='Over_Ratio', ascending=False)
teams = performance_leaders['Team']
x = range(len(teams))

# Initialize plot
fig, ax = plt.subplots(figsize=(14, 8))

# Scatter points
ax.scatter(performance_leaders['Over_Ratio'], x, color='green', label='Overperformer', s=100)
ax.scatter(performance_leaders['Under_Ratio'], [i + 0.2 for i in x], color='red', label='Underperformer', s=100)

# Add annotations
for i, row in performance_leaders.iterrows():
    ax.text(row['Over_Ratio'] + 0.01, i, f"{row['Top_Overperformer']} ({row['Over_Ratio']:.2f})", color='green', fontsize=9)
    ax.text(row['Under_Ratio'] + 0.01, i + 0.2, f"{row['Top_Underperformer']} ({row['Under_Ratio']:.2f})", color='red', fontsize=9)

# Formatting
ax.set_yticks([i + 0.1 for i in x])
ax.set_yticklabels(teams)
ax.set_xlabel('Goals / xG Ratio')
ax.set_title('Top Overperformer and Underperformer Forwards per Team')
ax.legend()
ax.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
